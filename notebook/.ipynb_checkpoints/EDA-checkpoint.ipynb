# 1. Introduction

This notebook will conduct Exploratory Data Analysis (EDA) on a heart disease dataset obtained from Kaggle (https://www.kaggle.com/datasets/krishujeniya/heart-diseae). The dataset contains medical attributes used to predict the likelihood of heart disease in patients. By analyzing the data, we aim to uncover insights into which indicators are most problematic and contribute significantly to the prediction of heart disease.

The dataset includes various features related to patient health, such as age, sex, chest pain type, resting blood pressure, cholesterol levels, and more. Through this analysis, we will identify key patterns and relationships among the variables that could help in assessing the risk of heart disease.

### Description of Variables

- **age**: Age of the patient (in years)
- **sex**: Sex of the patient (1 = male, 0 = female)
- **cp**: Chest pain type (0 = typical angina, 1 = atypical angina, 2 = non-anginal pain, 3 = asymptomatic)
- **trestbps**: Resting blood pressure (in mm Hg on admission to the hospital)
- **chol**: Serum cholesterol (in mg/dl)
- **fbs**: Fasting blood sugar > 120 mg/dl (1 = true; 0 = false)
- **restecg**: Resting electrocardiographic results (0 = normal, 1 = having ST-T wave abnormality, 2 = showing probable or definite left ventricular hypertrophy)
- **thalach**: Maximum heart rate achieved
- **exang**: Exercise-induced angina (1 = yes; 0 = no)
- **oldpeak**: ST depression induced by exercise relative to rest
- **slope**: Slope of the peak exercise ST segment (1 = upsloping, 2 = flat, 3 = downsloping)
- **ca**: Number of major vessels colored by fluoroscopy (0-3)
- **thal**: Thalassemia (1 = normal, 2 = fixed defect, 3 = reversible defect)
- **target**: Presence or absence of heart disease (1 = disease present, 0 = disease absent)

By exploring these variables, we will assess their impact on the likelihood of heart disease and identify the most critical indicators for effective prediction and diagnosis.

### Reference Values for Healthy Individuals

- **age**: Values vary with age. Compare against the general population distribution.
- **sex**: Not applicable (no "healthy" value; 1 = male, 0 = female).
- **cp**: 
  - 0 = typical angina (not indicative of health, generally should be absent).
- **trestbps**: 
  - Normal: < 120 mm Hg (systolic) and < 80 mm Hg (diastolic)
  - Higher values may indicate hypertension.
- **chol**: 
  - Normal: < 200 mg/dl
  - Higher values are generally considered elevated.
- **fbs**: 
  - Normal: 0 (Fasting blood sugar â‰¤ 120 mg/dl)
  - 1 = true may suggest possible diabetes.
- **restecg**: 
  - 0 = normal
- **thalach**: 
  - Normal: 60-100 bpm (varies, generally for resting heart rate).
- **exang**: 
  - 0 = no (no exercise-induced angina).
- **oldpeak**: 
  - Normal: 0 (no ST depression induced by exercise).
- **slope**: 
  - 1 = upsloping (generally healthy)
  - 2 = flat (may be normal but can indicate issues if associated with other symptoms)
  - 3 = downsloping (often indicates issues).
- **ca**: 
  - Normal: 0
  - Higher values may suggest abnormalities.
- **thal**: 
  - 1 = normal
- **target**: 
  - 0 = disease absent (healthy indicator).

These values serve as general guidelines and may vary based on different populations and medical standards.

Lets calculate the interqual (IQ) for each to see how many data points are outliers

```python
outliers_dict = {}  #Dictionary to save outliers index for later

# Function to detect the limits for each variable
def detect_outliers(df, feature, lower_percentile=0.1, upper_percentile=0.90):
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[feature] < lower_bound) | (df[feature] > upper_bound)].index

# Detect outliers and save index for later
for feature in continuos_vars:
    outliers = detect_outliers(df, feature)
    outliers_dict[feature] = outliers
    print(f'{feature}: {len(outliers)} outliers detected')

print("\nOutliers values for each feature:")
for feature, indices in outliers_dict.items():
    print(f'{feature}: {list(df[feature].iloc[indices.tolist()])}')
```

```python
df_no_outliers = df.copy()
index_to_drop = [lists for lists in outliers_dict.values() for item in lists]
#index_to_drop = [ itens for itens in list_index_to_drop for item in itens ]
list_index_to_drop = [list(lists.values) if len(lists)>0 else [] for lists in outliers_dict.values()]
index_to_drop = [ item for lists in list_index_to_drop for item in lists ]

df_no_outliers.drop(index = index_to_drop, inplace = True)
```

## EDA - Impact of each feature

Now we will take a further look at each feature, to see how it affects the other variable distribution.

```python
def chi_square_test(df, var1, var2, alpha=0.05):
    """
    Performs the Chi-Square test of independence between two categorical variables.
    
    Args:
        df (pd.DataFrame): The DataFrame containing the data.
        var1 (str): Name of the first categorical variable.
        var2 (str): Name of the second categorical variable.
        alpha (float): Significance level for the test. Default value is 0.05.
    
    Returns:
        dict: A dictionary with the test results.
    """
    # Create a contingency table (cross-tabulation) between the two variables
    contingency_table = pd.crosstab(df[var1], df[var2])
    
    # Perform the Chi-Square test of independence
    chi2, p_value, dof, expected = stats.chi2_contingency(contingency_table)
    
    # Interpret the p-value to determine if the relationship is statistically significant
    if p_value < alpha:
        result = "The variables are dependent (significant relationship)"
    else:
        result = "The variables are independent (no significant relationship)"
    
    return {
        'chi2_statistic': chi2,               # Chi-Square statistic
        'p_value': p_value,                   # p-value from the test
        'degrees_of_freedom': dof,            # Degrees of freedom
        'expected_frequencies': expected,     # Expected frequencies if variables were independent
        'result': result                      # Conclusion: dependent or independent
    }
```

```python
def anova_test(df, categorical_var, continuous_var):
    """
    Analyzes the impact of a categorical variable on a continuous variable.
    
    Parameters:
    df (pd.DataFrame): The dataframe containing the data.
    categorical_col (str): The name of the categorical column.
    continuous_col (str): The name of the continuous column.
    
    Returns:
    dict: A dictionary with the p-value and the result of the test (valid or not).
    """
    
    # Perform ANOVA
    model = ols(f'{continuous_var} ~ C({categorical_var})', data=df).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)
    
    # Extract the p-value from the ANOVA table
    p_value_anova = anova_table['PR(>F)'][0]
    
    # Determine if ANOVA assumptions are met
    valid_anova = p_value_anova < 0.05
    
    if valid_anova:
        result = 'ANOVA test is valid'
    else:
        result = 'ANOVA test is NOT valid'
        
    return {
        'anova_p_value': p_value_anova,
        'anova_result': result
    }

```

```python
def plot_histogram(df, var, categorial_vars):

    categorial_vars_aux = categorial_vars.copy()
    categorial_vars_aux.remove(var)

    num_vars = len(categorial_vars_aux)
    plt.figure(figsize=(12, 12))
    
    fig, axes = plt.subplots(nrows=(num_vars // 3) + 1, ncols=3, figsize=(15, 10))
    
    fig.suptitle(f'{var} impacts on categorical variables', fontsize=16, y=1.02)
    
    # Loop throw features
    for i, feature in enumerate(categorial_vars_aux):
        ax = axes[i // 3, i % 3]  # Seleciona a posiÃ§Ã£o correta do subplot
        sns.histplot(data=df, x=feature, ax=ax, hue=var, multiple="stack")
        ax.set_title(f'{feature}', fontsize=12)
        ax.set_ylabel('Count')
    
    # Remove empty subplots
    for j in range(i + 1, axes.shape[0] * axes.shape[1]):
        fig.delaxes(axes[j // 3, j % 3])
    
    plt.tight_layout()
    plt.show()
```

```python
def plot_boxplot(df, var, continuous_vars):
    
    num_vars = len(continuous_vars)
    plt.figure(figsize=(12, 12))
    
    fig, axes = plt.subplots(nrows=(num_vars // 3) + 1, ncols=3, figsize=(15, 10))
        

    
    fig.suptitle(f'{var} impacts on continuous variables', fontsize=16, y=1.02)
    
    # Loop throw features
    for i, feature in enumerate(continuos_vars):
        ax = axes[i // 3, i % 3]  # Seleciona a posiÃ§Ã£o correta do subplot
        ax.set_title(f'{feature}', fontsize=12)
        sns.boxplot(data=df, x=feature, y=var,orient="h", ax=ax)

    # Remove empty subplots
    for j in range(i + 1, axes.shape[0] * axes.shape[1]):
        fig.delaxes(axes[j // 3, j % 3])
    
    plt.tight_layout()
    plt.show()
```

# Conclusions from the EDA

This exploratory data analysis (EDA) of a heart disease dataset revealed several key insights and identified variables with significant associations with heart disease.

- **Age-Related Risk Factors**: Older age was associated with increased risk factors like elevated blood pressure and cholesterol, but decreased cardiovascular fitness as indicated by lower maximum heart rate and ST-segment depression.
Heart Disease and Age: Individuals aged 55-65 had a lower prevalence of heart disease and were less likely to experience typical angina.
- **Coronary Artery Disease Progression**: Coronary artery disease may worsen with age, as evidenced by the higher prevalence of no obstructed arteries (ca = 0) in older individuals with heart disease.
- **Correlations with Heart Disease**: Features with moderate correlation to heart disease (target) include chest pain type (cp), maximum heart rate (thalach), exercise-induced angina (exang), ST-segment depression (oldpeak), slope of the peak exercise ST segment (slope), number of major vessels colored by fluoroscopy (ca), and Thalassemia (thal). Features with low correlation include cholesterol (chol), fasting blood sugar (fbs), and resting electrocardiographic results (restecg).
- **Sex Differences**: Men were more likely to experience exercise-induced angina. Women showed a higher prevalence of heart disease and had higher cholesterol levels.
- **Chest Pain Analysis**: Typical angina (cp = 0) was associated with potential issues like flat slope (slope = 1) and Thalassemia types 2 or 3 (thal). Non-anginal pain (cp = 2) was more prevalent in individuals with heart disease.
- **Chest Pain Patterns**: Chest pain type 2 (non-anginal pain) was more commonly associated with heart disease, while chest pain type 3 (asymptomatic) was more prevalent in healthy individuals.
- **Fasting Blood Sugar**: FBS may not have a strong correlation with most variables but appears to be associated with higher age and elevated resting blood pressure.
- **Exang = 0 in Heart Disease**: A surprising finding was the higher prevalence of no angina (exang = 0) in individuals with heart disease compared to healthy individuals.
- **Slope Analysis**: Individuals with a severe slope (slope = 2) had a younger age distribution, lower ST-segment depression, and higher maximum heart rates, but also a higher prevalence of heart disease.
- **Number of Obstructed Arteries (ca)**: Higher values of ca were associated with increased risk factors like elevated resting blood pressure and older age. Interestingly, ca = 0 (no obstructions) was linked to a higher prevalence of features associated with heart disease, suggesting further investigation is needed.
- **Thalassemia (thal)**: Thal = 2 was associated with the highest risk of heart disease, while Thal = 2 and 3 were linked to a higher prevalence of heart disease overall.
